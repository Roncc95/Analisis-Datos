```python
def get_reduced_data(data,y_counts,ylab='City',n_min=20):
    y_val = y_counts[y_counts>n_min]
    data_val_df = data.loc[data[ylab].isin(y_val.index)]
    return data_val_df, y_val

y_counts = otus.City.value_counts()# Solo agarro datos con mas de 20 muestras segun el ciudad

n_min = 10
val_df, y_val = get_reduced_data(df,y_counts,'City',n_min)
```

Conjuntos de datos para entrenamiento y prueba

```python
y = np.array(val_df[y_names]).ravel() # ravel() me sirve para aplanar un array mutidimensional o un array unidimensional
X = np.array(val_df.drop(y_names, axis=1)).astype('float') # y es para el clima matriz de co-variables
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import preprocessing
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

Encoding de la variable de respuesta

```python
le = preprocessing.LabelEncoder()
le.fit(y)
#le.classes_
y_cat = le.transform(y)
#le.inverse_transform(y_cat[:5])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.3, random_state=0) # particion del conjunto de datos
```

```python
y_test
```

Pipeline
- estandarizacion
- reduccion de dimension
- clasificacion



```python
#Se realiza un grafica para ver cuantas componentes principales capturan la mayoria de la varianza de datos
pca = PCA().fit(X_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.grid(True)
```

```python
data = {'N': list(range(1,197)),
        'var': np.cumsum(pca.explained_variance_ratio_)}

df1 = pd.DataFrame(data)
df1["var"][30]
```

```python
pipe_steps = [('std', StandardScaler()), ('dim_red', PCA(n_components=30)), ('classifier', SVC(kernel='linear', C=1E6))]
pipe = Pipeline(steps=pipe_steps)
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
plt.style.use('default')
cm = metrics.confusion_matrix(y_test,y_pred)
disp1 = metrics.ConfusionMatrixDisplay(cm, display_labels=np.arange(len(le.classes_)))
fig, ax = plt.subplots(figsize=(7,7))

disp1.plot(ax=ax, colorbar=False)
# Adding custom colorbar
cax = fig.add_axes([ax.get_position().x1+0.01,ax.get_position().y0,0.02,ax.get_position().height])
plt.colorbar(disp1.im_,  cax=cax);
```
